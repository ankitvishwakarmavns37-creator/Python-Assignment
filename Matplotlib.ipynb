```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data={
    "Month" : ["jan", "feb", "mar", "apr", "may", "jun"],
    "Sales" : [3455,6456,7688,2345,6877,9765]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
plt.bar(df["Month"],df["Sales"])
```

```python
plt.figure(figsize=(6,3))
plt.bar(df["Month"],df["Sales"])
plt.show()
```

```python
colors=list(df["Sales"].map(lambda val : "red" if val<0 else "lightgreen"))
```

```python
colors
```

```python
plt.figure(figsize=(6,3))
plt.bar(df["Month"],df["Sales"] , color=colors)
plt.show()
```

```python
colors=list(df["Sales"].map(lambda val : "red" if val<0 else "lightgreen" if val<5000 else "lightpink"))
```

```python
plt.figure(figsize=(6,3))
plt.bar(df["Month"],df["Sales"] , color=colors)
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.bar(df["Month"],df["Sales"] , color=colors)
plt.title("Sales By Months")
plt.xlabel("Months Name")
plt.ylabel("Sales Amount in $USD")
plt.show()
```

```python
df=pd.read_excel(r"C:\Users\ankitvishwakarma\Downloads\financial_Sample.xlsx")
```

```python
df.shape
```

```python
df=df.bfill()
```

```python
df.head(2)
```

```python
pbc=df.groupby("Country").agg({"Profit" :"sum"}).reset_index()
```

```python
pbc["Country"]=pbc["Country"].map(lambda val:"USA" if val=="United States of America" else  val)
```

```python
pbc
```

```python
plt.figure(figsize=(6,3))
plt.pie(pbc["Profit"],labels=pbc["Country"] , autopct="%.2f%%")
plt.show()
```

```python
pbs=df.groupby("Segment")["Profit"].sum().reset_index()
```

```python
pbs
```

```python
plt.figure(figsize=(6,3))
plt.bar(pbs["Segment"], pbs ["Profit"])
plt.show()
```

```python
colors=list(pbs["Profit"].map(lambda val: "red" if val<0  else "orange" if val<3000000 else "green"))
```

```python

plt.figure(figsize=(6,3))
plt.bar(pbs["Segment"], pbs["Profit"], color=colors)
plt.show()
```
