```python
def convertVotes(val):
    val =str(val)
    if val[-1]=="K":
         return int(float(val[0: len(val)-1])*1000)
    elif val[-1]=="M":
         return int(float(val[0: len(val)-1])*1000000) 
    else:
        return int(val)
```

```python
convertVotes(3452)
```

```python
df.Votes = df.Votes.map( convertVotes )
```

```python
def convertDuration(val):
    st = val.split()
    if len(st)==2:
        
        return int( st[0][0: len(st[0])-1])*60 + int( st[1][0: len(st[1])-1])
    elif val[-1]=="H":
        return int( val[0: len(val)-1])*60
    elif val [-1]=="M":
        return int(val[0:len(val)-1])
    else:
        return convertDuration(df.Duration.mode()[0])
```
