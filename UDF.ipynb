```python
# *a will behave like a tuple 
def add(*a):
    return sum(a)
print(add(10,20,30,40,50,60))
```

```python
# **a will behave like dict
def fun(**a):
    return a
print( fun( name="tej" ) )
```

```python
# recursion function()
def greet():
    for i in range(5):
        print("ankit")
    
greet()
```

```python
def fact(num):
    f=1
    for i in range(1,num+1):
        f=f*i
    return f
print("Factorial=",fact(5))
```

```python
# use recursion () this function is call by self its call recursion
def fact(num):
    if num==1:
        return 1
    else:
        return num * fact(num-1)
print("Factorial=",fact(5))
```

```python
# wap to add all natura number from 1 to n
def add(num):
    if num==1:
        return 1
    else:
        return num+add(num-1)
print("Add Natural Num =",add(10))
```
