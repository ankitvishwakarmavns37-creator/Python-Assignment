### Set Programming Questions
### Basic

### 1. Create a set and add elements dynamically.

```python
my_set = set()

my_set.add(10)
my_set.add(20)
my_set.add(30)

print("Set after adding elements:", my_set)

n = int(input("How many elements you want to add: "))

for i in range(n):
    val = int(input("Enter element: "))
    my_set.add(val)

print("Final Set:", my_set)
```

### 2. Find the union and intersection of two sets

```python
set1={1,2,3,4}
set2={3,4,5,6}
union_set=set1|set2
intersection_set=set1&set2
print("set1 :",set1)
print("set1 :",set2)
print("union :",union_set)
print("intersection :",intersection_set)
```

 ### 3. Remove duplicate elements from a list using a set.

### 6. Find common elements in two lists using sets.

```python
list1=[1,2,3,4,5]
list2=[3,4,5,6,7]
set1=set(list1)
set2=set(list2)
common_ele=set1&set2
print("list1",list1)
print("list2",list2)
print("Common Element: ",list(common_ele))
```

### 11. Find missing numbers from 1 to n using sets.

```python
nums=[1,2,4,5,7]
n=6
full_set=set(range(1,n+1))
nums_set=set(nums)
missing=full_set-nums_set
print("Missing n number :",missing)
```

### 13. Convert a set of strings into uppercase.

```python
set1={"ankit","tej","sanju"}
upper_case={item.upper()for item in set1} 

print("upper case :",upper_case)
```

### 14. Identify unique vowels in a given string using a set

```python
text = "programming in python"
vowels = "aeiou"

result = {char for char in text.lower() if char in vowels}

print("Unique vowels:", result)
```

### 15. Find elements that appear only once in a list

```python
my_list = [1, 2, 2, 3, 4, 4, 5]

unique_once = [x for x in my_list if my_list.count(x) == 1]

print("Elements appearing only once:", unique_once)
```
