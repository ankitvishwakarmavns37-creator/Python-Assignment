# 8. Write a program to print all prime numbers between 1 and 100

```python
for num in range(2, 101):
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num,end=" ")

        
```

# 16. Write a program to keep taking input from the user until the user enters 0.

```python
count = 0
total = 0
num = 1

while num != 0:
    num = int(input("Enter a number (0 to stop): "))
    if num != 0:
        total += num
        count += 1

print("Count =", count)
print("Sum =", total)
```

# 18. Write a program to check whether a number is a palindrome.

```python
num = int(input("Enter a number: "))
original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10

if original == reverse:
    print("Palindrome number")
else:
    print("Not a palindrome number")
```

# 19. Write a program to print the Fibonacci series up to n terms.

```python
num=int(input("enter a number"))
a=0
b=1
i=1
while i<=num:
    print(a)
    c=a+b
    a=b
    b=c
    i+=1
```

# 20. Write a program to implement a number guessing game using a while loop.

```python
import random
secret_number = random.randint(1, 6)

guess = None
attempts = 0

print("Welcome to Number Guessing Game!")
print("Guess a number between 1 and 6")

while guess != secret_number:
    guess = int(input("Enter your guess: "))
    attempts += 1

    if guess < secret_number:
        print("Too low! Try again.")
    elif guess > secret_number:
        print("Too high! Try again.")
    else:
        print(" Congratulations! You guessed it right!")
        print("Total attempts:", attempts)
```

# 23. Write a program to print all Armstrong numbers between 1 and 1000

```python

for num in range(1, 1001):
    temp = num
    sum_digits = 0
    digits = len(str(num))

    while temp > 0:
        digit = temp % 10
        sum_digits += digit ** digits
        temp //= 10

    if sum_digits == num:
        print(num)
```


# 24. Write a program to simulate an ATM menu using a while loop

```python
balance = 5000

while True:
    print("1 Balance  2 Deposit  3 Withdraw  4 Exit")
    ch = int(input("Choice: "))

    if ch == 1:
        print(balance)
    elif ch == 2:
        balance += int(input("Amount: "))
    elif ch == 3:
        balance -= int(input("Amount: "))
    elif ch == 4:
        break
```

# 25. Write a program to find the GCD of two numbers using loops

```python
a=int(input("enter the first number"))
b= int(input("enter the second number"))
i=1
gcd=1
while i<=min(a,b):
    if a%i==0 and b%i==0:
        gcd=i
    i+=1
print("gcd=",gcd)
    
```
