# 1. Write a program to count the number of vowels in a string.

```python
st=input("enter the string")
vowel="aeiouAEIOU"
count=0
for ch in st:
    if ch in vowel:
        count+=1
print("enter the string",count)
```

# 2. Reverse a string without using built-in functions

```python
st=input("enter the string")
rev=""
for ch in st:
    rev=ch+rev
print("Reverse string:=",rev)
```

# 3. Check whether a string is a palindrome

```python
st=input("enter the string:=")
rev=""
for ch in st:
    rev=ch+rev

if st==rev:
    print("palindrom number")
else:
    print("not palindrom")
```

# 4. Count uppercase and lowercase letters in a string.

```python
st=input("enter the string:=")
upper=0
lower=0
for ch in st:
    if ch>="A" and ch<="Z":
        upper+=1
    elif ch>="a" and ch<="z":
        lower+=1
print("UPPERCASE letter:=",upper)
print("lowercase letter:=",lower)
    
```

# 6. Find the frequency of each character in a string.

```python
st=input("enter the number:=")
fre={}
for ch in st:
    if ch in fre:
        fre[ch]+=1
    else:
        fre[ch]=1
print("character frequency:")
for key in fre:
    print(key,":", fre[key])
```

# 8. Find the first non-repeating character in a string.

```python
s = input("Enter a string: ")

freq = {}

for ch in s:
    freq[ch] = freq.get(ch, 0) + 1

found = False
for ch in s:
    if freq[ch] == 1:
        print("First non-repeating character:", ch)
        found = True
        break

if not found:
    print("No non-repeating character found")
```

# 9. Check if two strings are anagrams.

```python
st1=input("enter the first string")
st2=input("enter the second string")
if sorted(st1)==sorted(st2):
    print("anagrams")
else:
    print("not anagrams")
```

# 12. Compress a string like "aaabbc" → "a3b2c1".

```python
st="aaabbc"
temp=""
count=1
i=0
while i<len(st)-1:
    if st[i]==st[i-1]:
        count+=1
    else:
        temp=temp + st[i-1] + str(count)
        count=1
    i+=1
temp=temp+st[i-1]+str(count)
print(temp)
```

# 13. Count words, characters, and digits in a string.

```python
text="python programming language 2026"
word_count=len(text.split())
char_count=len(text)
digit_count=sum (1 for ch  in text if ch.isdigit())
print(word_count)
print(char_count)
print(digit_count)
```

# 15. Find all substrings of a given string.

```python
text="abc"
substring=[]
for i in range (len(text)):
    for j in range (i+1, len(text)+1):
        substring.append (text[i:j])
print(substring)
    
```
