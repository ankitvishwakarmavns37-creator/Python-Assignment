# 14. Write a program to flatten a nested list

```python
nested_list = [1, [2, 3], [4, [5, 6]], 7]
flatten_list=[]
for item in nested_list:
    if type (item )== list:
        for i in item:
            flatten_list .append(i)
    else:
            flatten_list.append(item)
print(flatten_list)
```

# 30. Write a program to find repeated elements in a tuple

```python
tup = (10, 20, 30, 20, 40, 30)

repeated = []

for item in tup:
    if tup.count(item) > 1 and item not in repeated:
        repeated.append(item)

print("Repeated elements:", tuple(repeated))
```
